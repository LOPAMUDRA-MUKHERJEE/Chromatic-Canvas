```python
#Total number of orders made from each state
df.groupby('State')['Order ID'].count()
```

```python
#Total quantities sold from each category
df.groupby('Category')['Quantity'].sum()
```

```python
#Number of orders of each category from each state
df.groupby(['State','Category'])['Order ID'].count()
```

```python
#Average of sales and profit made by each category
df.groupby('Category')[['Sales','Profit']].mean()
```

```python
#Total of sales and profit made by each category
df.groupby('Category')[['Sales','Profit']].sum()
```

```python
#Processing Tome of each order
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Processing Time'] = df['Ship Date'] - df['Order Date']
df['Processing Time']
```

```python
#Total sales & profit made in each year
df['Year'] = df['Order Date'].dt.year
df.groupby('Year')[['Sales','Profit']].sum()
```
